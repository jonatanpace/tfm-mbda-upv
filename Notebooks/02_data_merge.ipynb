```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
robos = pd.read_csv('../Datasets/Clean/robos_all_years.csv')
dist_a_comisaria = pd.read_csv('../Datasets/Clean/distancia_entre_delito_y_comisaria_clean.csv')
weather_data_clean = pd.read_csv('../Datasets/Clean/weather_data_clean.csv')
```

```python
robos.head()
```

```python
dist_a_comisaria.head()
```

```python
weather_data_clean.head()
```

```python
robos_enriquecido = robos.merge(
    right = dist_a_comisaria, 
    left_on = 'id-mapa', 
    right_on = 'id_delito', 
    how = 'left').drop(columns = 'id_delito').rename(columns={}).merge(
        right = weather_data_clean,
        left_on = ['fecha', 'franja'],
        right_on = ['date', 'hour'],
        how = 'left'
            ).drop(columns = ['date', 'hour'])
robos_enriquecido
```

```python
# sanity check
robos_enriquecido[[
    'anio', 
    'fecha', 
    'franja', 
    'time', 
    'w_temperatura'	,
    'w_humedad_relativa',	
    'w_precipitaciones',
    'w_velocidad_del_viento',
    'w_presion_del_aire',
    'w_codigo_del_tiempo',
    ]].sample(10)
```

```python
robos_enriquecido.to_csv('../Datasets/Clean/robos_enriquecido.csv', index = False)
```
